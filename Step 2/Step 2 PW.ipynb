```python
import pandapower as pp
import pandapower.networks as nw
import pandas as pd
from pandapower.timeseries.data_sources.frame_data import DFData
from pandapower.timeseries.output_writer import OutputWriter
from pandapower.control import ConstControl
from pandapower.timeseries.run_time_series import run_timeseries

#create empty net
net = nw.create_cigre_network_mv()
```

```python
load_shape = pd.read_excel('Loadshape_step_2.xlsx', usecols=[1])
```

```python
num_hours = len(load_shape)
```

```python
net.load
```

```python
load_profiles = pd.DataFrame(index=range(24), columns=net.load.index)
```

```python
for t in range(num_hours):
    load_profiles.loc[t] = net.load['p_mw'].values * load_shape.iloc[t,0]
```

```python
for h in range(num_hours):
    print(load_shape.iloc[h])
```

```python
ds = DFData(load_profiles)
```

```python
for load_idx in net.load.index:
    ConstControl(net, element='load', variable='p_mw', element_index=load_idx, data_source=ds, profile_name=load_idx)
```

```python
time_steps = range(num_hours)
```

```python
ow = OutputWriter(net, time_steps=time_steps, output_path=None)
ow.log_variable('res_bus', 'vm_pu')
ow.log_variable('res_line', 'loading_percent')
```

```python
run_timeseries(net, time_steps=time_steps)
```

```python
bus_voltages = {}
line_loadings = {}
```

```python
for t in time_steps:
    bus_voltages[t] = net.res_bus['vm_pu'].values
    line_loadings[t] = net.res_line['loading_percent'].values
```

```python
df_bus_voltages = pd.DataFrame(bus_voltages)
df_line_loadings = pd.DataFrame(line_loadings)
```

```python
with pd.ExcelWriter('time_series_results.xlsx') as writer:
    df_bus_voltages.to_excel(writer, sheet_name='Bus_Voltages')
    df_line_loadings.to_excel(writer, sheet_name='Line_Loadings')

print("Results successfully written to time_series_results.xlsx")
```
